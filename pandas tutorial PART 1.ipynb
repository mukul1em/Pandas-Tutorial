## Filter the data

```python
data=pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv') #comma seperated format
```

```python
data.head()
```

1 if we want the rows which has duration greater than 175

```python
booleans=[]
for length in data.duration:
    if length >= 175:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[:20]
```

How it works?

we have the list of booleans containing True,False
when we pass the list in the data

say if 1st is data[True] then it will keep first row

if it is data[False] then it will remove that row in order of boolean list

```python
data[booleans] # we have the movies which are greater than 175
```

```python
# another method of filtering
```

```python
data[data.duration>=175]
```

# Groupby function

```python
# in imdb dataset i want to check for different genre what is the mean or max, any aggergate duration 
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv')
data.head()
```

```python
data.groupby('genre').duration.mean()
```

```python
data.groupby('genre').duration.min()
```

```python
data.groupby('genre').duration.agg(['min','max','mean','median','count'])
```

```python
data.groupby('genre').agg(['min','max','median','count','mean'])
```
